```python
from Repair.repair_algorithm import RepairAlgorithm
import numpy as np 
import pandas as pd
import Scenarios.scenario_types.ScenarioConfig as sc
from Scenarios.scenario_types.AnomalyRate import AnomalyRateScenario

import matplotlib.pyplot as plt

cols = [5]

repair_alg = RepairAlgorithm("cdrec" , cols)
anomaly = "shift"
data = "msd.csv"
injected_scenario = AnomalyRateScenario(data,cols,anomaly_type = 
                                                      anomaly)


estimator = repair_alg.estimator
name ,train,test =list(injected_scenario.name_train_test_iter)[3]
    
test , klass = test["injected"],test["class"]
test_reduced1 = estimator.reduce(test.values,1)
pd.DataFrame(test_reduced1).iloc[:,cols].plot(title=1)
test_reduced2 = estimator.reduce(test.values,2)
pd.DataFrame(test_reduced2).iloc[:,cols].plot(title=2)
test_reduced3 = estimator.reduce(test.values,3)
pd.DataFrame(test_reduced3).iloc[:,cols].plot(title=3)


pd.DataFrame(test-test_reduced1).iloc[:,cols].plot(title=11)

pd.DataFrame(test-test_reduced2).iloc[:,cols].plot(title=12)

pd.DataFrame(test-test_reduced3).iloc[:,cols].plot(title=13)
np.linalg.matrix_rank(estimator.transform_matrix)
```
